# Understanding Lambda functions

```python
lambda x: x

# In the example above, the expression is composed of:

# The keyword: lambda
# A bound variable: x
# A body: x
```

```python
# You can apply the function above to an argument by surrounding the function and its argument with parentheses:

(lambda x: x+1)(2)
```

Reduction is a lambda calculus strategy to compute the value of the expression. It consists of substituting the argument 2 for x:

(lambda x: x + 1)(2) = lambda 2: 2 + 1
                     = 2 + 1
                     = 3

```python
# Because a lambda function is an expression, it can be named. 
# Therefore you could write the previous code as follows.

add_one = lambda x: x+1
```

```python
add_one(2)
```

```python
# Multiple argument
full_name = lambda fn, ln: f'Full Name: {fn.title()} {ln.title()}'
full_name('Debadatta', 'Bhattacharjee')
```

```python
# Odd or even
# 3%2 == False
oddEve = lambda x: x%2 and 'odd' or 'even'
```

```python
oddEve(3)
```
